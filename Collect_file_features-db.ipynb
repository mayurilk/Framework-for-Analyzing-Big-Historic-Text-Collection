```python
import os
import glob
import json
import shutil
import pickle
```

```python
def get_feature_vectors(folder_name):
    files_to_features =[]
    sub_folders = [f1 for f1 in glob.glob(folder_name + os.sep + '*')]
    for sf in sub_folders:
        files = [f2 for f2 in glob.glob(sf + os.sep + '*')]
        for f in files:
            with open(f,'rb') as fp:
                data = pickle.load(fp)
                if len(data)>0:
                    for k,v in data.items():
                        f_name = os.path.basename(f).split('features_')[1] + "_" + k
                        files_to_features.append((f_name,v))
    return files_to_features
```

```python
folder_name = "/Users/dan/Documents/01_DataScience/02_IIT_CS522_2017/Processing/FEATURES"
file_features = get_feature_vectors(folder_name)
```
